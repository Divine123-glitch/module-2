# Question 1.


## A.
<br>
<br>

Find the mean, median and mode of the following students scores in a class mathematics test:
$$
scores =  5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20
$$

<br>
<br>

The mean of the scores above is given by:
<br>

$$
\frac{1}{n}\sum^n_{i = 1}{x_i}, \text{ where }x_i  \in scores
$$

<br>
<br>

Using the formula above, we denote that:
$$
mean = \frac{5 + 8 + 10 + 12 + 12 + 12 + 13 + 14 + 14 + 15 + 16 + 18 + 18 + 20 + 20}{15} = 13.8
$$

<br>
<br>

The median can be gotten by sorting our scores and finding the mid score. Our scores has been sorted, the next step is to find the mid score(s)
$$\therefore median = 14$$

<br>
<br>

The mode is the value with the highest frequency, so
$$ Mode = 12$$

<br>
<br>
<br>

## B.
<br>
<br>
If the lowest score (5) was wrongly recorded and should have been 10, how would the mean change?
<br>

$$
Original Scores = 5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20
$$

<br>

$$
New Scores = 10, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20
$$

<br>

$$
New mean = \frac{10 + 8 + 10 + 12 + 12 + 12 + 13 + 14 + 14 + 15 + 16 + 18 + 18 + 20 + 20}{15} = 14.13$$

<br>

The difference in mean is ( New mean - Original mean ):
$$14.13 - 13.8 = 0.33$$

<br>

** The mean increases because the corrected score (10) is higher than the original score (5), increasing the total sum by 5, which raises the average by $  \approx 0.3333 $. **
<br>

** The median remains 14, as the correction does not affect the middle score in the ordered list. **
<br>

** The mode is still 12. Changing the score of 5 to 10 increases the frequency of 10 from 1 to 2, but 12 still has the highest frequency (3). Thus, the mode is unchanged. **
